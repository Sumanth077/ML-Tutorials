```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a Sample data to be plotted

x = np.linspace(0, 5)

y = [x, x**2, np.sqrt(x), x**3]

#Define the title of the subplots

title = ['x', 'x^2', 'sqrt(x)', 'x^3']
```

```python
# Define a figure 
plt.figure(figsize=(8, 8))

#Plot the Subplots using subplot method
for i in range(4):
    axes = plt.subplot(2, 2, i+1)
    axes.plot(x, y[i], color= np.random.rand(3,))
    axes.set_title(title[i])
```
